### Dicounted Cash Flow

It's what's available to the company after they cover capital expenditures.

But what's available to the company to grow the business or maybe pay out as dividents or pay off debts or whatever they elect to do with it.

Works well with blue-chip companies:
- Coca Cola
- Walmart
- McDonalds

Don't use it with insurance companies or banks r investment banks (JP Morgan) (coz their cashflow might be a bit skewed)

### Divident Valuations
The stock pays a lot of consistent divident, and if those dividents are mostly tied to the profit of the company.

So if they pay out a large percent of their profits.

For this model to performs well:
- The dividents should be the primary source of shareholder wealth in that particular investment.

E.g. if the stock doesn't move much but it consistently pays out a good divident.

DCF - the cash available to the company.
DV - what's actually being paid out to investors.

Candidates:
- Real estate
- investment trust
- high-yielding ETF

### Asset Based Models:
The theory here is that the fair value of any company's assets minus the value of the comapny's liabilities.



It's **not** as simple as taking the book value of the company, whwich is on the balance sheet and it's calculated in a very similar way. Because in that they're using the depreciated vale of an asset, and that might not really be the fair value of the actual asset. So they're not really comparable.

"Floor value of the company".

Works well for compnies that are close to liquidation or if a company consists of primarity tangible assets that have a market-ready price.

E.g.

- companies that drill for oil (oil company)
Hypothetically they've done surveys, they have an idea of how much oil is in the ground, they publish that info.

We could very quickly find out the price of oil, so we do a simple calculation how much oil do they have access to, how much oil is right now and you could come up with a rough estimate of what the value of that asset would be. Subtract what they owe, here you go.

- Macy's (liquidation company)
