```python
data = yf.Ticker("IBM")
# data.info
```

Shares Outstanding:
how many shares is the company divided into

```python
shares_outstanding = data.info["sharesOutstanding"]
shares_outstanding
```

EPS: Earning Per Share

Profit one share make for an entire year

```python
net_income = data.info["netIncomeToCommon"]
eps = net_income / shares_outstanding
print(net_income, eps)
```

Diluted EPS: sometimes companies distribute their share to their employees and they don't write them down on sheets. So in reality they create shares

More calculated, use this one

P/E
E.g. P/E = 100, which means
for every $100 I use to buy this share, I can expect a dollar in the following year

```python
current_price = data.info["currentPrice"]
pe = current_price / eps
print(current_price, pe)
```

## Exploring data

```python
data_10 = data.history(period="10y")
data_5 = data.history(period="5y")
```

```python
data_10.Dividends
```



It's 2.745% for 29/05/2022

```python
dvd_avg = data_10.Dividends.sum()/10
dvd_avg
```

```python
risk_free_rate = 0.02745
dvd_avg / (1 + risk_free_rate)**3
```

The result means, if we use \\$4.87  to purchase the 10-year national bond today, it will increase to \\$5.28 in 3 years due to the interest.

We can repeat this calculation for the next 10 years, but an easier way is to use a for-loop.

```python
dvd_total = 0
for i in range(1,11):
    dvd_total = dvd_total + dvd_avg/(1+risk_free_rate)**i
print(dvd_total)
```

So, we have an estimation for the next 10-years dividends

```python
import requests
from bs4 import BeautifulSoup

URL = 'https://www.macrotrends.net/stocks/charts/IBM/ibm/price-book'
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

bvps = {}

tbody = soup.find(id="style-1").find("tbody")
for tr in tbody.findAll("tr"):
    trs = tr.findAll("td")
    if ("12-31" not in trs[0].text): continue
    year = int(trs[0].text.split("-")[0])
    price = float(trs[1].text.strip())
    print(year, "    ", price)
    bvps[year] = price
```

```python
diff = max(bvps.keys()) - min(bvps.keys())
growth_rate = max(bvps.keys()) / min(bvps.keys())
avg_growth_rate = growth_rate**(1/diff)
avg_growth_rate
```

```python
bvps_pred = bvps[max(bvps.keys())] * avg_growth_rate
bvps_pred
```

```python
ibm_intrinsic = bvps_pred + dvd_total
ibm_intrinsic
```
